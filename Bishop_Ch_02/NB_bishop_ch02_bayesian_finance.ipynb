## 1. Bayesian A/B Testing


```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Fintech: two landing page variants
np.random.seed(42)
n_A, conv_A = 1000, 45  # Control: 4.5%
n_B, conv_B = 1000, 58  # Treatment: 5.8%

# Beta posteriors (uniform prior)
alpha_A, beta_A = 1 + conv_A, 1 + n_A - conv_A
alpha_B, beta_B = 1 + conv_B, 1 + n_B - conv_B

x = np.linspace(0, 0.12, 500)
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, stats.beta.pdf(x, alpha_A, beta_A), 'b-', lw=2, label=f'A: {conv_A}/{n_A}')
ax.plot(x, stats.beta.pdf(x, alpha_B, beta_B), 'r-', lw=2, label=f'B: {conv_B}/{n_B}')
ax.legend(); ax.set_xlabel('Conversion rate'); ax.set_ylabel('Posterior density')
ax.set_title('Bayesian A/B Test: Posterior Distributions')
plt.show()

# P(B > A) via Monte Carlo
samples_A = np.random.beta(alpha_A, beta_A, 100000)
samples_B = np.random.beta(alpha_B, beta_B, 100000)
prob_B_better = np.mean(samples_B > samples_A)
print(f'P(B > A) = {prob_B_better:.1%}')
```

## 2. Credit Default Screening (Base Rate Fallacy)


```python
# Test with 95% sensitivity, 90% specificity
# Default rate = 2%
sensitivity, specificity, prevalence = 0.95, 0.90, 0.02

p_pos_default = sensitivity
p_pos_no_default = 1 - specificity
p_pos = p_pos_default * prevalence + p_pos_no_default * (1 - prevalence)
p_default_given_pos = (p_pos_default * prevalence) / p_pos

print(f'P(default | flagged) = {p_default_given_pos:.1%}')
print(f'Despite 95% sensitivity, only {p_default_given_pos:.1%} of flagged loans actually default!')
print(f'False positive rate: {1 - p_default_given_pos:.1%}')
```
